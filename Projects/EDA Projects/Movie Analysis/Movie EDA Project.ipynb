```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'movies.csv')
```

```python
# Now let's take a look at the data

df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
sns.regplot(x="gross", y="budget", data=df)
```

### Conclusion : strong positive correlation between a movie's budget and its gross earnings. This means that, in general, movies with higher budgets tend to make more money at the box office.

```python
sns.regplot(x="score", y="gross", data=df)
```

### conclusion : weak positive correlation between movie score and gross. As movie score increases, gross also slightly increases. However, the correlation is not very strong, suggesting that movie score is not a major factor in determining gross.

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

# Final Conclusions

### 1. Trends Over the Years:
The number of movies released each year can show fluctuations, with certain years having significantly higher numbers of releases. This could be due to various factors such as economic conditions, technological advancements, and changes in audience preferences.

### 2. Correlation Analysis:
Budget vs Gross Earnings: There is generally a positive correlation between a movie's budget and its gross earnings. Higher budget movies tend to earn more, likely due to better production quality, marketing, and star power.
Votes and Score: Movies with higher votes often have higher scores, indicating that popular movies tend to be well-received by audiences.

### Impact of Ratings:
Gross Earnings: Movies rated PG-13 and R tend to have higher gross earnings compared to other ratings. This could be due to these movies targeting a broader or more specific audience.
Scores: Movies rated R and PG-13 also tend to have higher scores, suggesting that these movies are not only popular but also critically acclaimed.

### 4. Genre Analysis:
Gross Earnings: Action, Adventure, and Sci-Fi genres tend to have higher gross earnings. These genres often involve high-budget productions that attract large audiences.
Scores: Drama and Documentary genres often receive higher scores, indicating that these genres are well-appreciated for their storytelling and depth.

### 5. Director and Star Influence:
Certain directors and stars have a significant impact on a movie's success. Directors like Steven Spielberg, Christopher Nolan, and stars like Tom Hanks and Leonardo DiCaprio are associated with higher grossing and higher scoring movies.

### 6. Company Influence:
Movies produced by major companies like Warner Bros., Universal Pictures, and Disney tend to have higher budgets and gross earnings. These companies have more resources for production and marketing, contributing to the success of their movies.

### 7. Country of Production:
The majority of high-grossing movies come from the United States, reflecting the dominance of Hollywood in the global film industry. However, movies from the United Kingdom and other countries also show strong performance, particularly in specific genres.

# Final Remarks:
The success of a movie is influenced by a combination of factors including budget, rating, genre, director, star power, production company, and country of origin.
Understanding these factors can help stakeholders in the film industry make informed decisions regarding production, marketing, and distribution to maximize both critical and commercial success.
