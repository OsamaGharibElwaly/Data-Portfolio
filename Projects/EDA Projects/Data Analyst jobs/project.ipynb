```python
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('DataAnalyst.csv')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop('Unnamed: 0', axis = 1 , inplace = True)
```

```python
# Create a dictionary to map old column names to new ones
column_mapping = {
    'Job Title': 'job_title',
    'Salary Estimate': 'salary_estimate',
    'Job Description': 'job_description',
    'Rating': 'rating',
    'Company Name': 'company_name',
    'Location': 'location',
    'Headquarters': 'headquarters',
    'Size': 'size',
    'Founded': 'founded',
    'Type of ownership': 'type_of_ownership',
    'Industry': 'industry',
    'Sector': 'sector',
    'Revenue': 'revenue',
    'Competitors': 'competitors',
    'Easy Apply': 'easy_apply'
}

# Rename columns using the dictionary
df.rename(columns=column_mapping, inplace=True)
```

```python
df.columns
```

```python
df['job_title'][0:20]
```

```python
plt.figure(figsize=(12, 8))

# Find the most frequent job titles
top_job_titles = df['job_title'].value_counts().head(10).index.tolist()

# Filter the dataset for the most frequent job titles
subset_job_titles = df[df['job_title'].isin(top_job_titles)]

# Extract specific columns from the subset
job_titles = subset_job_titles['job_title']
salary_estimates = subset_job_titles['salary_estimate']

plt.figure(figsize=(20, 20))
plt.bar(job_titles, salary_estimates, color='green')
plt.xlabel('Job Title')
plt.ylabel('Salary Estimate')
plt.title('Job Title and Salary Estimate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Sr data analyst is the biggest salary

```python
# Distribution of company ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'].dropna(), bins=20)
plt.xlabel('Company Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Company Ratings for Data Analyst Jobs')
plt.show()
```

```python
# Sample 50 rows randomly from the DataFrame
sampled_data = df.sample(n=50, random_state=42)

# Sort the sampled data by 'Salary Estimate' in ascending order
sampled_data_sorted = sampled_data.sort_values(by='salary_estimate')

# Create the bar chart
plt.figure(figsize=(20, 20))
plt.bar(sampled_data['company_name'], sampled_data['salary_estimate'], color='skyblue')
plt.xlabel('Company Name')
plt.ylabel('Salary Estimate')
plt.title('Random Sample of 50 Companies with Salaries for Data Analyst Jobs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Countplot of job locations
plt.figure(figsize=(50, 50))
sns.countplot(y='location', data=df, order=df['location'].value_counts().index)
plt.xlabel('Number of Job Postings')
plt.ylabel('Location')
plt.title('Distribution of Data Analyst Jobs by Location')
plt.show()
```

# New York the city where the most data analyst work in

```python
# Convert 'Revenue' column to numerical format
revenue_mapping = {
    'Unknown / Non-Applicable': 0,
    '$0 to $1 million (USD)': 1,
    '$1 to $5 million (USD)': 2,
    '$5 to $10 million (USD)': 3,
    '$10 to $25 million (USD)': 4,
    '$25 to $50 million (USD)': 5,
    '$50 to $100 million (USD)': 6,
    '$100 to $500 million (USD)': 7,
    '$500 million to $1 billion (USD)': 8,
    '$1 to $2 billion (USD)': 9,
    '$2 to $5 billion (USD)': 10,
    '$5 to $10 billion (USD)': 11,
    '$10+ billion (USD)': 12,
    ' -1': 0  # Replace any other non-numeric values with 0
}
df['revenue'] = df['revenue'].map(revenue_mapping)

# Sample 50 rows randomly from the DataFrame
sampled_data = df.sample(n=50, random_state=42)

# Sort the sampled data by 'Founded' in ascending order
sampled_data_sorted = sampled_data.sort_values(by='founded')

# Create the bar chart
plt.figure(figsize=(12, 8))
plt.bar(sampled_data_sorted['founded'], sampled_data_sorted['revenue'], color='skyblue')
plt.xlabel('Founded')
plt.ylabel('Revenue')
plt.title('Founded vs Revenue for Random Sample of 50 Companies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Most companies were founded in the end of 20th century and the beginnings of 21th century
